```python
class Miniso:
    def __init__(self, id):
        self.id = id
        self.cart = {}
        print("Welcome to Miniso!!!")

    def addItem(self, *items):
        self.items = items
        for item in self.items:
            if item in self.cart.keys():
                print(f"{item} is already in the cart.")
            else:
                self.cart[item] = 1


    def addQuantity(self, item, num = 1):
        self.item = item
        self.num = num

        self.cart[self.item] = self.num
        #print(self.cart)

    def printDetails(self):
        if len(self.cart) == 0:
            print("Your cart is empty.")
        else:
            print(f"Cart ID: {self.id}")
            for key, val in self.cart.items():
                  print(f"{key}: {val}x")

        if len(self.cart) >= 10:
            print("Congratulations! You have won a voucher!")


m1 = Miniso('1012')
print('1===============================')
m1.addItem('Hello Kitty Comb','Hulk Plushy','Hello Kitty Comb')
m1.addQuantity('Hello Kitty Comb', 2)
m1.addQuantity('Hulk Plushy')
print('2===============================')
m1.printDetails()
m2 = Miniso('1013')
print('3===============================')
m2.printDetails()
m2.addItem('Coca Cola Sipper','Marvel Notebook','Thor Headphone','Grizzly Mug','Penguin Plushy')
m2.addQuantity('Coca Cola Sipper', 3)
m2.addQuantity('Penguin Plushy', 2)
m2.addItem('Ice Bear Pouch')
m2.printDetails()
m2.addItem('Marvel Notebook','Neck Pillow')
m2.printDetails()
```
